```python
poverty_file='/content/drive/MyDrive/DSGP/poverty/Povertylines.xlsx'
demo_file='/content/drive/MyDrive/DSGP/poverty/demograpic_district wise.xlsx'
```

```python
def load_data(poverty_file , demo_file):
    # Load poverty lines
    poverty_df = pd.read_excel(poverty_file)
    # Calculate average poverty line
    poverty_df['average_poverty_line'] = poverty_df.iloc[:, 1:].mean(axis=1)

    # Load demographic data
    demo_df = pd.read_excel(demo_file)
    # Aggregate population by district using the latest projection (PPROJ_22)
    district_pop = demo_df.groupby('DISTRICT_N')['PPROJ_22'].sum().reset_index()
    district_pop.rename(columns={'DISTRICT_N': 'District', 'PPROJ_22': 'Population'}, inplace=True)

    # Merge datasets on District
    merged_df = pd.merge(district_pop, poverty_df[['District', 'average_poverty_line']], on='District', how='inner')

    return merged_df
```
